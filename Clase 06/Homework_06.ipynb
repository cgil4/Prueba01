1) Cargar el dataset "Emisiones_CO2.csv" provisto en la clase 2 en un Dataframe de Pandas, quitar los registros que contengan valores faltantes e implementar una nueva columna, que contenga el resultado de una función Hash aplicada sobre el campo "Código de País" y se denomine "Clave_Hash".<br>
Consideraciones: Se puede utilizar la función provista.


```python
def hash_function(key):
    return sum(index * ord(character) for index, character in enumerate(repr(key), start=1))
```

```python
hash('a') #usar funcion hash arriba para que sean valores unicos
```

```python
print(hash_function('gato'))
print(hash_function('tapo'))
```

```python
import pandas as pd
```

```python
df = pd.read_csv('Emisiones_CO2.csv',sep='|',decimal=',', encoding='latin-1', thousands ='.')
```

```python
# desde aqui empiezo yo
df.dropna(inplace=True)  #para guardar los cambios en el mismo archivo sin los valores nulos
```

```python
df.rename(columns={'Código de país' : 'Cod_pais','Nombre del país' : 'Nom_pais',
                'CO2 (kt)' : 'Emisiones_co2','CO2 per cápita (toneladas métricas)' : 'Emico2_percapita'}, inplace= True)
                # inplace guarda cambios en el mismo archivo
```

```python
# df['Cod_pais'] # para traer una columna por su nombre
df['Clave_Hash'] = df['Cod_pais'].apply(hash_function)  # crea la columna y agrega los valores al cod de pais con la funcion hash
```
